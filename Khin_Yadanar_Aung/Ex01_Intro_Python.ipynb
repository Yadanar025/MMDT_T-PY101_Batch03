# Ex2 : Convert Currency.

# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
apple_price = 8
orange_price = 4
mango_price = 10
apple_qty = input ("Enter the quantity of apples: ")
orange_qty = input ("Enter the quantity of oranges: ")
mango_qty = input ("Enter the quantity of mangoes: ")
total_cost = (int(apple_qty) * apple_price) + (int(orange_qty) * orange_price) + (int(mango_qty) * mango_price)
print("The total cost of {}apple {}orange {}mango is: {}".format(apple_qty, orange_qty, mango_qty, total_cost))
```

# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
second = input("Enter time in seconds: ")
minute = int(second) // 60
minute_mod = int(second) % 60
print("{} seconds is {} minutes and {} seconds".format(second, minute, minute_mod))
```

# Ex 10: Using Math Module

Write a program that asks the user for a number and then prints out the sine, cosine, and tangent of that number.

```python
import math

num = float(input("Enter a number:"))
sine_value = math.sin(num)
print(sine_value)
```

```python
import math

number = float(input("Enter a number:"))
cosine_value = math.cos(number)
print("The cosine value of {} is: {}".format(number, cosine_value))
```

```python
import math

number = float(input("Enter a number:"))
tangent_value = math.tan(number)
print("The tangent value of {} is: {}".format(number, tangent_value))
```
